```python
df['ID'].to_numpy()
```

```python
df['score']=np.arange(len(df['ID'].to_numpy()))
df['score'].to_numpy()
```

```python
add=[a+b for a , b in zip(df['ID'],df['score']) ]
print("vector addition")
print(add)
print("vector subtraction")
sub=[a-b for a , b in zip(df['ID'],df['score']) ]
print(sub)
```

```python
dot_product=np.dot(df['ID'],df['score'])
print("dot product")
print(dot_product)
```

```python
Weights= np.array([0.3, 0.7])
df[['ID', 'score']].dot(Weights).to_numpy()
```
